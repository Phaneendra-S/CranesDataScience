```python
encoder=ce.OneHotEncoder(cols='State')
df=encoder.fit_transform(df)
df.head()
```

```python
#split the dataset into input features and target column
x=df.drop('Profit',axis=1)
y=df['Profit']
y=pd.DataFrame(y)
```

```python
#Scaling the input feature data

from sklearn.preprocessing import MinMaxScaler

scaler=MinMaxScaler()
x_scaled=scaler.fit_transform(x)
x_scaled=pd.DataFrame(x_scaled)
```

```python
#divide into train test dataset

from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.2,random_state=2)
```

```python
#Apply ML algo into training data
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred=pd.DataFrame(y_pred,index=y_test.index,columns=['y_pred_profit'])
y_pred
```

```python
#model evaluation
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
r2_score(y_test,y_pred)
```
