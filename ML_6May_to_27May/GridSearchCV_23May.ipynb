**GridSearchCV**
- Grid Search with Cross Validation
- Its a brute force method to find the best combination of hyperparameters for a machine learning model.

**Working**
1. Define a grid of possible hyper parameters
2. Training and validating the model for every combination using cross-validation
3. Picking the combination that gives the best performance in validation data

**Working**
- Define the model (e.g., SVM, Random Forest, etc.)
- Specify hyperparameters and their possible values in a dictionary

**Run GridSearch CV**
1. Trains the model on every combo using cross-validation(e.g.5-fold)
2. Calculates performance metrics(e.g. accuracy, F1 score)
3. Chooses the best combination based on average validation score
