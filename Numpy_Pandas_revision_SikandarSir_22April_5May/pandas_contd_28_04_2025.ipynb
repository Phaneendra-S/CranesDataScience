```python
data = {'Name': ['Abhishek', 'Suyash','Neelam','Yuvika'], 'CGPA': [7.3, 8, 8.6,9.2]}
df_sec_a = pd.DataFrame(data)
df_sec_a['CGPA'].mean()
```

```python
data2 = {'Name': ['Arvind', 'Mitansh','Ayush','Diksha','Aastha'], 'CGPA': [7.9, 7.8, 8.4, 9.3, 9.0]}
df_sec_b = pd.DataFrame(data2)
df_sec_b['CGPA'].mean()
```

```python
df_concat=pd.concat([df_sec_a,df_sec_b],axis=0,ignore_index=True)
df_concat
```

```python
df_concat_avg=df_concat['CGPA'].mean()
df_concat_avg
```
