**matplotlib and seaborn**

-matplotlib    
-its basic graphs pie,bar,line      
-datasets and arrays     
-its organized and more functional     

-seaborn   
-its advanced package built on top of matplotlib    
-mainly used for statistics visualization    
-it works entire dataset    
-it contains lot of inbuilt themes   

different types of line chart      

line chart      
it used to show the trends over period of time      
it is works finr with continuous data and time series analysis      
      
bar chart     
it used to compare the data across different categories
the values can be categorical     
   
pie chart       
it is used to find the proportion of each category        
the values can be categorical     

scatter plot       
it is used for relation between 2 variables
the values are continuous and bivariate    

histogram     
it is used for data distribution       
the valuse should be continuous and univariate           

boxplot         
it is used for identifying in outliers in data set       


```python
import matplotlib.pyplot as plt
```

```python
x=[-4,-6,1,5,8]
y=[15,10,8,6,2]
fig=plt.figure()
plt.plot(x,y)
plt.xlabel("x list")
plt.ylabel("y list")
plt.title("line chart")
plt.show()
```
