```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Create a figure with a specific size
plt.figure(figsize=(2,2))

# Plot some data
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]
plt.plot(x, y)

# Add titles and labels
plt.title('Sample Plot with Custom Figure Size')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Show the plot
plt.show()
```

```python
data=np.random.randint(150, size=(150))
print(data)
plt.figure(figsize=(2,2))
plt.hist(data, bins=8, color='skyblue', edgecolor='black')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```

```python
plt.figure(figsize=(2,2))
c=[90,85,45,70,60]
label=['DS','Embed','Vlsi','iot','java']
plt.pie(c,labels=label,autopct='%1.0f%%')
plt.show()
```

```python
x=np.random.rand(50)
y=np.random.rand(50)

plt.figure(figsize=(2,2))
plt.scatter(x,y,color='green',marker='.',label='Random points')

plt.title('Scatter plot example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.show()
```

```python
data=np.random.randint(100,size=(100))
print(data)

plt.figure(figsize=(5,5))
plt.boxplot(data)

plt.title('Box plot with outliers')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.show()
```

```python
x=np.linspace(0,10,100)
y=np.sin(x)

plt.plot(x,y,color='red', linestyle='--', linewidth=2,label='Sine Wave')
plt.fill_between(x,y,color='lightcoral', alpha=0.3)
plt.grid(True,linestyle='--')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Customized sine wave plot')
plt.legend()

plt.show()
```

```python
# car.cyl.value_counts().plot(kind='pie')
def show_counts(pct, all_vals):
    absolute = int(round(pct / 100 * sum(all_vals)))
    return f'{absolute}'

cyl_counts = car['cyl'].value_counts().sort_index() 
values = cyl_counts.values 
label = [f'{cyl} Cylinders' for cyl in cyl_counts.index] 

print(car.cyl.value_counts())
plt.pie(car.cyl.value_counts(),labels=label,autopct=lambda pct: show_counts(pct, values))
plt.show()
```
