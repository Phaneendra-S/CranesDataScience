**apply() method**
- applies a function to every row in the dataframe

```python
def to_uppercase(column):
    return column.upper()
```

```python
df['Capitalised_title']=df['original_title'].apply(to_uppercase)
df
```

```python
df['lowercase_title']=df.apply(lambda df:df['original_title'].lower(),axis=1)
df
```

```python
df.dtypes
```

```python
#create a new column called profit using apply() method
def profit(revenue,budget):
    return revenue-budget

df['profit']=df.apply(lambda df:profit(df['revenue'],df['budget']), axis=1)
df
```

```python
df['loss']=df.apply(lambda df:(df['budget']-df['revenue']), axis=1)
df
```

**create a new column "Old_or_New" such that if the year is greater than 2015 print new film else old film**

```python
df['Year']=df['new_release_date'].dt.year
df.head()
```

```python
df['Old or New']=df.apply(lambda df:('New' if df['Year']>2015 else 'Old'),axis=1)
df
```
